```python
import os
import pandas as pd
import matplotlib.pyplot as plt

file_path = os.path.join(os.getcwd(),"..", "data", "clean_data.csv")
df = pd.read_csv(file_path)
```

```python
df = df[df["Regions"] != "The Netherlands"] # exclude the netherlands
df = df[df["Travel motives"] == "Total"] # only look at travel motives row which says total
df = df[df["Travel modes"] != "Total"] # delete the row where travel modes is total
df["Travel modes"] = df["Travel modes"].replace({"Passenger car (driver)": "Car", "Passenger car (passenger)": "Car"}) 
# merge car driver and car passenger into one travelmode
```

```python
df = df[df['Year'].between(2018.0, 2023.0)]
avg_trips_by_mode = df.groupby(['Regions','Travel modes'])['Average number of trips per person per year'].mean().reset_index()
# groups data by regions and travel modes and calculates the mean number of trips per person per year
```

```python
# sums the total trips per pers per year for each region
totals = avg_trips_by_mode.groupby('Regions')['Average number of trips per person per year'].sum().reset_index()
totals = totals.rename(columns={'Average number of trips per person per year': 'totaltrips'})
avg_trips_by_mode = avg_trips_by_mode.merge(totals, on='Regions', how='left')
# calculates the share in percentages of each travel mode in each region
avg_trips_by_mode['percentage'] = (avg_trips_by_mode['Average number of trips per person per year'] / avg_trips_by_mode['totaltrips']) * 100
print(avg_trips_by_mode)
```

```python
modes = avg_trips_by_mode['Travel modes'].unique()
for mode in modes:
    subset = avg_trips_by_mode[avg_trips_by_mode['Travel modes'] == mode]
    plt.figure(figsize=(10,5))
    plt.bar(subset['Regions'], subset['percentage'])
    plt.title(f"{mode} - Share of total trips per province")
    plt.xlabel("Province")
    plt.ylabel("Percentage of total trips (%)")
    plt.xticks(rotation=45)
    ymax = subset['percentage'].max() * 1.1  
    plt.ylim(0, ymax)

    plt.tight_layout()
    plt.show()
```

```python
pivot_df = avg_trips_by_mode.pivot_table(index='Regions',columns='Travel modes', values='percentage',fill_value=0)

pivot_df = pivot_df.sort_index()

pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title("Share of travel modes per province (stacked, 100%)")
plt.xlabel("Province")
plt.ylabel("Percentage of total trips (%)")
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.legend(title="Travel mode", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
